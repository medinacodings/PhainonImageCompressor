2. Okay, this time, i'm going to build the script by loading the photo and error handling to counter when unable to load or could not find any image.

```python
# import relevant libraries
import cv2
import numpy as np

# load the image from the folder
def modify_photo(image_path):
    # Use the passed image_path instead of hardcoding
    image = cv2.imread(image_path)
    return image

# error handling when the image couldn't be loaded
if image is None:
    print(f"Error: Unable to find or load image: {image_path}")
```

3. Now we have the photo placed into the script, let's try to convert the photo into grayscale a.k.a black and white.

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

4. Let's apply the gaussian blur to smooth out the pixel noise. I determine the blur intensity to be 15, 15.

```python
blurred = cv2.GaussianBlur(gray, (15, 15), 0)
```

5. Enhance the contrast

```python
enhanced_gray = cv2.equalizeHist(gray)
```

6. Let's brighten the image mathematically, so I created an array of the same size filled with value 50.

```python
bright_matrix = np.ones(image.shape, dtype="uint8") * 50
brightened = cv2.add(image, bright_matrix)
```

7. Okay, the scripts are done. Let's save the results back to the folder.

```python
cv2.imwrite('phainon_gray_pic.jpg', gray)
cv2.imwrite('phainon_blurred_pic.jpg', blurred)
cv2.imwrite('phainon_contrast_pic.jpg', enhanced_gray)
cv2.imwrite('phainon_bright_pic.jpg', brightened)
```

8. Alright, all done! let's use Phainon's photo to see the script works.

```python
print('Phainon modified photos are successfully done!')

# let's execute to see the magic!
if __name__ == '__main__':
    image_path = '/content/phainon_splash-art.png'
    modified_image = modify_photo(image_path)
```
